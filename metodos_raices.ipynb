```python
import matplotlib.pyplot as plt
from scipy.optimize import newton
import numpy as np


def f(x):
    return x**5 + 5.25*x**4 + 4.125*x**3 - 9.125*x**2 - 14.625*x - 5.625

x = np.linspace(-4, 2, 500)
plt.plot(x, f(x))
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.title('Gr√°fica de f(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()
plt.show()

```

```python
def f(x):
    return x**5 + 5.25*x**4 + 4.125*x**3 - 9.125*x**2 - 14.625*x - 5.625

initial_guesses = (-1.20,0,5)
roots = []

for x0 in initial_guesses:
    try:
        r = newton(f, x0)
        r = np.round(r, 6)
        if not any(np.isclose(r,roots)):
            roots.append(r)
    except:
        pass

print("Roots found:", roots)
for r in roots: 
    print(r)
```
